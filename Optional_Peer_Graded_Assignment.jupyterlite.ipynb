## Task 3: Use the appropriate tests to answer the questions provided.

#### 1.- Is there a significant difference in median value of houses bounded by the Charles river
#### or not? (T-test for independent samples)

###### H_0: There is no a significant difference in median values between CHAS and MEDV
###### H_1: There is a significant difference in median values between CHAS and MEDV

```python


scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == '1']['MEDV'],
                      boston_df[boston_df['CHAS'] == '0']['MEDV'], equal_var = True)
```

#### 2.-Is there a difference in Median values of houses (MEDV) for each proportion of owner
#### occupied units built prior to 1940 (AGE)? (ANOVA)

###### H_0: There is no difference in median values of houses MEDV and AGE
###### H_1: There is difference in median values of houses MEDV and AGE

```python

#group1 = boston_df[boston_df['age_group'] == '50 years and younger']['MEDV']
#group2 = boston_df[boston_df['age_group'] == 'between 50 and 70 years']['MEDV']
#group3 = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
from statsmodels.formula.api import ols
lm = ols('MEDV ~ age_group', data = boston_df).fit()
table = sm.stats.anova_lm(lm, typ=2)
print(table)
```

##### There is a significant difference between groups acording with p-value

#### 3.-Can we conclude that there is no relationship between Nitric oxide concentrations and

#### 4.-What is the impact of an additional weighted distance to the five Boston employment
#### centres on the median value of owner occupied homes? (Regression analysis)

##### H_0: Additional weighted distance to the five Boston employment centres has impact 
##### H_1: Additional weighted distance to the five Boston employment centres has no impact 

```python

X = boston_df['DIS']
y = boston_df['MEDV']
X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```
